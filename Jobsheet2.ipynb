```python
import cv2 as cv
from google.colab.patches import cv_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('/content/drive/MyDrive/images/images.jpeg')
img2 = cv.cvtColor(img,cv.COLOR_BGR2RGB)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.imshow(img2)
```

2. Show images in Red-Green and Green-Blue channels only!

```python
rg_img = img2.copy()
rg_img[:,:,2] = 0

gb_img = img2.copy()
gb_img[:,:,0] = 0

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(rg_img)

plt.subplot(1,2,2)
plt.imshow(gb_img)
plt.show()
```

3. Show image row 10-100, column 10-100!

```python
cropped_img = img2[10:100, 10:100]

plt.imshow(cropped_img)
plt.show
```

4. Show image rows 5-30, all columns, channel Green only!

```python
green_channel = img2[5:30, :, 1]

plt.imshow(green_channel)
plt.show()
```

5. Create 5 boxes of different sizes and colors in one image. it is recommended to use
random numbers!

```python
img_boxes = img2.copy()
height, width, _ = img2.shape

for _ in range(5):

  x1, y1 = np.random.randint(0,width), np.random.randint(0, height)
  box_width, box_height = np.random.randint(20,100), np.random.randint(20,100)
  color = tuple(np.random.randint(0,256,3).tolist())

  cv.rectangle(img_boxes, (x1,y1), (x1 + box_width, y1 + box_height),color,thickness=2)

plt.imshow(img_boxes)
plt.show()
```

6. Show the image upside down!

```python
flipped_img = np.flipud(img2)

plt.imshow(flipped_img)
plt.show
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# Load a sample image, e.g., "lena.png"
img_path = '/content/drive/MyDrive/images/Cute-Cat.jpg'  # Replace with the correct path
img = cv.imread(img_path)
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Extract the blue channel
blue_channel = img[:, :, 0]

# Create a rectangle in the lower-left corner of the blue channel
start_point = (10, blue_channel.shape[0] - 50)  # Bottom left corner
end_point = (100, blue_channel.shape[0] - 10)   # Define width and height
color = (255)  # Since we are in a single channel, use a grayscale value
cv.rectangle(blue_channel, start_point, end_point, color, thickness=-1)  # Filled rectangle

# Display the modified image
plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel with Rectangle')
plt.show()
```
